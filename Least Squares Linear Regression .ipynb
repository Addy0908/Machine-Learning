```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
#import dataset
df = pd.read_csv('ls.csv')
print('The dataset used to train is\n\n',df)
```

```python
#generating X and Y
X = df['X_col'].values
Y = df['Y_col'].values
```

```python
# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)
 
# Total number of values
n = len(X)
```

```python
# Using the formula to calculate 'm' and 'c'
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
    m = numer / denom
    c = mean_y - (m * mean_x)
     
# Printing coefficients
print("Coefficients")
print(m, c)
```

```python
##The above coefficients are our slope and intercept values respectively
```

```python
# Plotting Values and Regression Line
 
max_x = np.max(X) + 5
min_x = np.min(X) - 5
 
# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = c + m * x
 
# Ploting Line
plt.plot(x, y, color='#58b970', label='Regression Line')
# Ploting Scatter Points
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')
 
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Calculating Root Mean Squares Error
rmse = 0
for i in range(n):
    y_pred = c + m * X[i]
    rmse += (Y[i] - y_pred) ** 2
rmse = np.sqrt(rmse/n)
print("RMSE")
print(rmse)
```

```python
# Calculating R2 Score
ss_tot = 0
ss_res = 0
for i in range(n):
    y_pred = c + m * X[i]
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_res/ss_tot)
print("R2 Score")
print(r2)
```

```python
#As you can see our R-squared value is quite close to 1, this denotes that our model is doing good and can be used for further predictions.
```
